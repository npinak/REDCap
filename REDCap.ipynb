
#**REDCap

###- Used REDCap to create a baseline data entry form to capture patient data for a hypothetical study comparing 2 analgesics. 
###- Added validation to fields such as “Height” to make sure only an integer value can be added.
###- Used Branching logic in fields such as “Gender” to ensure that the “Pregnancy Status” question is not displayed if the participant selects “Male”. 
###- Added section headers to the form such as “Confounding Factors” to ensure that the data form looks neat and orderly.
